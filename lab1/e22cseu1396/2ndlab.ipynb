```python
import cv2
import numpy as np
from google.colab import drive
import matplotlib.pyplot as plt

image_path = "/content/drive/MyDrive/example.jpg"
image = cv2.imread(image_path)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

def display_images(images, titles):
    plt.figure(figsize=(15, 5))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i+1)
        plt.imshow(img)
        plt.title(title)
        plt.axis('off')
    plt.show()

display_images([image_rgb], ["Original Image"])

linear_resized = cv2.resize(image, (300, 300), interpolation=cv2.INTER_LINEAR)

nearest_resized = cv2.resize(image, (300, 300), interpolation=cv2.INTER_NEAREST)

polynomial_resized = cv2.resize(image, (300, 300), interpolation=cv2.INTER_CUBIC)

display_images(
    [cv2.cvtColor(linear_resized, cv2.COLOR_BGR2RGB),
     cv2.cvtColor(nearest_resized, cv2.COLOR_BGR2RGB),
     cv2.cvtColor(polynomial_resized, cv2.COLOR_BGR2RGB)],
    ["Linear Resized", "Nearest Neighbors Resized", "Polynomial (Cubic) Resized"]
)

box_blur = cv2.blur(image, (5, 5))

gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)

adaptive_blur = cv2.bilateralFilter(image, 9, 75, 75)

display_images(
    [cv2.cvtColor(box_blur, cv2.COLOR_BGR2RGB),
     cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB),
     cv2.cvtColor(adaptive_blur, cv2.COLOR_BGR2RGB)],
    ["Box Blurring", "Gaussian Blurring", "Adaptive (Bilateral) Blurring"]
)
```

```python
import cv2
import numpy as np
from google.colab import drive
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
from sklearn.tree import DecisionTreeClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import seaborn as sns

drive.mount('/content/drive')

from tensorflow.keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.reshape(-1, 28 * 28) / 255.0  # Flatten images
X_test = X_test.reshape(-1, 28 * 28) / 255.0

y_train_onehot = to_categorical(y_train)
y_test_onehot = to_categorical(y_test)

kf = KFold(n_splits=5, shuffle=True, random_state=42)
decision_tree_accuracies = []

def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    conf_matrix = confusion_matrix(y_test, y_pred)
    return accuracy, precision, recall, f1, conf_matrix

for train_idx, val_idx in kf.split(X_train):
    X_train_fold, X_val_fold = X_train[train_idx], X_train[val_idx]
    y_train_fold, y_val_fold = y_train[train_idx], y_train[val_idx]
    model = DecisionTreeClassifier()
    accuracy, precision, recall, f1, conf_matrix = evaluate_model(model, X_train_fold, X_val_fold, y_train_fold, y_val_fold)
    decision_tree_accuracies.append(accuracy)

ann_model = Sequential([
    Flatten(input_shape=(28 * 28,)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

ann_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
ann_model.fit(X_train, y_train_onehot, epochs=10, batch_size=32, verbose=1)

ann_metrics = ann_model.evaluate(X_test, y_test_onehot, verbose=0)

print("Decision Tree Accuracies (K-Fold):", decision_tree_accuracies)
print(f"Decision Tree Average Accuracy: {np.mean(decision_tree_accuracies):.4f}")

print(f"ANN Accuracy: {ann_metrics[1]:.4f}")

y_pred_ann = np.argmax(ann_model.predict(X_test), axis=1)
conf_matrix_ann = confusion_matrix(y_test, y_pred_ann)

plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_ann, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - ANN")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```
